```python
import pandas as pd
import numpy as np
import re
```

```python
# UTF-8 Windows-1252 Latin-1
autos = pd.read_csv("autos.csv",encoding="Latin-1")
```

```python
autos.info()
autos.head()
```

```python
columns = autos.columns
print(type(columns.to_numpy()))
print(columns.to_numpy())
```

```python
change={'yearOfRegistration':'registration_year',
'monthOfRegistration' : 'registration_month',
'notRepairedDamage' : 'unrepaired_damage',
'dateCreated': 'ad_created'
}
autos.rename(change,axis=1,inplace=True)
columns = autos.columns
print(autos.columns)
```

```python
def convert(name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower()
new_columns = []
for element in columns:
    element = convert(element)
    new_columns.append(element)
autos.columns = new_columns
columns = autos.columns
print(autos.columns)
```

```python
# store price as numeric
autos["price"] = autos["price"].str.replace("$","").str.replace(",","").astype(float)
```

```python
# store odometer as numeric
autos["odometer"] = autos["odometer"].str.replace("km","").str.replace(",","").astype(float)
```

```python
brand = autos["brand"].value_counts(normalize=True)
brand_selected = brand[brand>0.05]
selected_list = brand_selected.index
mean_price = {}
mean_odo = {}
for i in selected_list:
    mean_price[i] = autos.loc[autos["brand"]==i,"price"].mean()
    mean_odo[i] = autos.loc[autos["brand"]==i,"odometer_km"].mean()
print(mean_price)
print(mean_odo)
```

```python
bmp_series = pd.Series(mean_price)
odo_series = pd.Series(mean_odo)
print(bmp_series)
print(odo_series)
```

```python
df_bmp = pd.DataFrame(bmp_series, columns=["mean_price"])
df_odo = pd.DataFrame(odo_series, columns=["mean_odo"])
print(df_bmp)
print(df_odo)
```

```python
# import
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline
```

## input and explore data

```python
# input data: notice that the raw data doesn't have col_index
cols = ['symboling', 'normalized_losses', 'make', 'fuel_type', 'aspiration', 'num_of_doors', 'body_style', 
        'drive_wheels', 'engine_location', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 
        'engine_type', 'num_of_cylinders', 'engine_size', 'fuel_system', 'bore', 'stroke', 'compression_rate', 
        'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']
cars_df = pd.read_csv("imports-85.data",names=cols)
cars_df.head()
```

```python
#cars_df.info()
```

## Data cleaning

```python
#only select numeric columns relating to price
selected_cols = ['normalized_losses', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_size', 
                 'bore', 'stroke', 'compression_rate', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']
numeric_cars = cars_df[selected_cols]
```

```python
#numeric_cars.info()
```

```python
# replace ? with nan
numeric_cars = numeric_cars.replace('?', np.nan)
# convert all attributes' type to float
numeric_cars = numeric_cars.astype("float")
# check null-value
#numeric_cars.isnull().sum()
```

```python
# Remove all rows with missing `price` values.
numeric_cars = numeric_cars.dropna(subset=["price"])
#numeric_cars.isnull().sum()
```

```python
# Replace missing values in other columns using column means.
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
#numeric_cars.isnull().sum()
```

```python
# normalize all columns except "price"
price_col = numeric_cars["price"]
# function: x= (x-min)/max-min range is [0,1]
numeric_cars = (numeric_cars-numeric_cars.min()) / (numeric_cars.max()-numeric_cars.min())
numeric_cars["price"]=price_col
numeric_cars.head()
```

## Univariate Model
- only use 1 attribute to predict

```python
def knn_handout(train_col,target,df):
    np.random.seed(1)
    # Randomize the order in df
    df = df.reindex(np.random.permutation(df.index))
    # handout validation(2 fold) 50% : 50%
    mid = int(len(df)/2)
    train_df = df[0:mid]
    test_df = df[mid:]
    knn = KNeighborsRegressor()
    knn.fit(train_df[[train_col]],train_df[target])
    prediction = knn.predict(test_df[[train_col]])
    mse = mean_squared_error(test_df[target],prediction)
    rmse = mse ** 0.5
    return rmse
# use handout validation to analyze each single column except "price"
rmse_result = {}
train_cols = numeric_cars.columns.drop("price")
for c in train_cols:
    rmse_result[c] = knn_handout(c,"price",numeric_cars)
rmse_result_series = pd.Series(rmse_result)
rmse_result_series = rmse_result_series.sort_values()
#print(rmse_result_series)
# create a scatter to display
fig,ax = plt.subplots()
ax.scatter(rmse_result_series.index,rmse_result_series)
ax.set_xticklabels(rmse_result_series.index,rotation=90)
ax.set_title("Univariate Model(by handout validation)")
ax.set_xlabel("Column Name")
ax.set_ylabel("RMSE")
plt.show()
```

### we use k=5 as default above, now try to use k in [1,3,5,7,9]

```python
def knn_handout_k(train_col,target,df,ks):
    np.random.seed(1)
    # Randomize the order in df
    df = df.reindex(np.random.permutation(df.index))
    # handout validation(2 fold) 50% : 50%
    mid = int(len(df)/2)
    train_df = df[0:mid]
    test_df = df[mid:]
    
    # create a dictionary to store k-results for each k
    k_dic = {}
    for k in ks:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[train_col]],train_df[target])
        prediction = knn.predict(test_df[[train_col]])
        mse = mean_squared_error(test_df[target],prediction)
        rmse = mse ** 0.5
        k_dic[k] = rmse
    return k_dic
# for each col generate k_dic by using knn_handout_k
# in this example, set k = [1,3,5,7,9]
rmse_per_col = {}
k = [1,3,5,7,9]
for col in numeric_cars.columns.drop("price"):
    rmse_per_col[col] = knn_handout_k(col,"price",numeric_cars,k)
#print(rmse_per_col)
cols_labels=[]
for col,k in rmse_per_col.items():
    x = list(k.keys())
    y = list(k.values())
    plt.plot(x,y)
    cols_labels.append(col)
plt.xlabel('Value of K')
plt.ylabel("RMSE")
plt.legend(labels=cols_labels,bbox_to_anchor=(1.05, 0), loc=3, borderaxespad=0)
```

## Multivariate Model
- Use multi-attributes(features) to calculate Euclidean distance

```python
# Compute average RMSE across different k values for each feature
feature_avg_rmse = {}
for col,k in rmse_per_col.items():
    avg_rmse = np.mean(list(k.values()))
    feature_avg_rmse[col] = avg_rmse
avg_rmse_series = pd.Series(feature_avg_rmse).sort_values()
feature_rank = avg_rmse_series.index
print(avg_rmse_series)
print(feature_rank)
```

### Ranks of features is generated above, now explore different k influencing x-best-features

```python
def knn_handout_k_multi(train_cols,target,df,ks):
    np.random.seed(1)
    # Randomize the order in df
    df = df.reindex(np.random.permutation(df.index))
    # handout validation(2 fold) 50% : 50%
    mid = int(len(df)/2)
    train_df = df[0:mid]
    test_df = df[mid:]
    
    # create a dictionary to store k-results for each k
    k_dic = {}
    for k in ks:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[train_cols],train_df[target])
        prediction = knn.predict(test_df[train_cols])
        mse = mean_squared_error(test_df[target],prediction)
        rmse = mse ** 0.5
        k_dic[k] = rmse
    return k_dic
# for x-best generate k_dic by using knn_handout_k_multi
n_best_dic = {}
k = [x for x in range(1,26)]
n_best_dic["1 best feature"] = knn_handout_k(feature_rank[0],"price",numeric_cars,k)
#print(n_best_dic)
for i in range(2,6):
    n_best_dic[str(i)+" best feature"] = knn_handout_k_multi(feature_rank[:i+1],"price",numeric_cars,k)
#print(n_best_dic)
cols_labels=[]
for col,k in n_best_dic.items():
    x = list(k.keys())
    y = list(k.values())
    plt.plot(x,y)
    cols_labels.append(col)
plt.xlabel('Value of K')
plt.ylabel("RMSE")
plt.legend(labels=cols_labels)
plt.show()
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score, KFold
num_folds = [3, 5, 7, 9, 10, 11, 13, 15, 17, 19, 21, 23]
folds_dic = {}
stds_dic = {}
for fold in num_folds:
    kf = KFold(fold, shuffle=True, random_state=1)
    model = KNeighborsRegressor()
    mses = cross_val_score(model, numeric_cars[[feature_rank[0]]], numeric_cars["price"], scoring="neg_mean_squared_error", cv=kf)
    rmses = np.sqrt(np.absolute(mses))
    avg_rmse = np.mean(rmses)
    std_rmse = np.std(rmses)
    folds_dic[fold] = avg_rmse
    stds_dic[fold] = std_rmse
fig = plt.figure(figsize=(10,10))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
ax1.plot(list(folds_dic.keys()),list(folds_dic.values()),label="avg RMSE",color='red')
ax2.plot(list(stds_dic.keys()),list(stds_dic.values()),label="avg STD",color='green')
ax1.legend()
ax2.legend()
ax2.set_xlabel("Number of Plots")
ax1.set_ylabel("RMSE")
ax2.set_ylabel("RMSE")
ax1.set_title("Comparisons of RMSE and STD when Plots change")
plt.show()
    
```
